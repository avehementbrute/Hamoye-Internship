```python
#importing dependencies
import pandas as pd ,numpy as np 

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
%autosave 1
```

```python
train_csv = pd.read_csv('D:/Dataset/planets_dataset/planet/planet/train_classes.csv')
train_csv
```

```python
labels = set()

def split_the_tags(tags):
    for tag in tags.split():
        labels.add(tag)
        
train_csv_copy = train_csv.copy(deep=True)
train_csv_copy['tags'].apply(split_the_tags)

print(labels,end='')
```

```python
for tag in labels:
    train_csv_copy[tag] = train_csv_copy['tags'].apply(lambda x: 1 if tag in x.split() else 0)
    
train_csv_copy
```

```python
train_csv_copy['image_name'] = train_csv_copy['image_name'].apply(lambda x: f'{x}.jpg')
train_csv_copy
```

```python
#importing tensorflow libraries 
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, TensorBoard
from tensorflow.keras.layers import Dense, BatchNormalization, Conv2D, MaxPooling2D, Dropout, Flatten
```

```python
#for the analysis we only need columns after index 2
columns = list(train_csv_copy.columns[2:])
print(columns,end='')
```

```python
#function for f1 scoring
def f1_score_fun(y_true, y_pred, beta = 2, epsilon = 1e-4):
    
    beta_squared = beta**2
    
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(tf.greater(tf.cast(y_pred, tf.float32), tf.constant(0.5)), tf.float32)
    
    tp = tf.reduce_sum(y_true * y_pred, axis = 1)
    fp = tf.reduce_sum(y_pred, axis = 1) - tp
    fn = tf.reduce_sum(y_true, axis = 1) - tp
    
    precision = tp/(tp+fp+epsilon)
    recall = tp/(tp+fn+epsilon)
    
    fb = (1+beta_squared)*precision*recall / (beta_squared*precision+recall+epsilon)
    return fb
```

```python
#function for accuracy for multi_label classification
def accuracy_fun(y_true, y_pred, epsilon = 1e-4):
    
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(tf.greater(tf.cast(y_pred, tf.float32), tf.constant(0.5)), tf.float32)
    
    tp = tf.reduce_sum(y_true * y_pred, axis = 1)
    fp = tf.reduce_sum(y_pred, axis = 1) - tp
    fn = tf.reduce_sum(y_true, axis = 1) - tp
    
    y_true = tf.cast(y_true, tf.bool)
    y_pred = tf.cast(y_pred, tf.bool)
        
    tn = tf.reduce_sum(tf.cast(tf.logical_not(y_true), tf.float32) * tf.cast(tf.logical_not(y_pred), tf.float32), 
                       axis = 1)
    return (tp+tn)/(tp+tn+fp+fn+epsilon)
```

```python
#defining our model using a function build_model()
def build_model():
    model = Sequential()
    model.add(BatchNormalization(input_shape=(128, 128, 3)))
    model.add(Conv2D(32, kernel_size=(3, 3), padding='same', activation='relu'))
    model.add(Conv2D(32, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.2))

    model.add(Conv2D(64, kernel_size=(3, 3), padding='same', activation='relu'))
    model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.2))

    model.add(Conv2D(128, kernel_size=(3, 3), padding='same', activation='relu'))
    model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.2))

    model.add(Conv2D(256, kernel_size=(3, 3), padding='same', activation='relu'))
    model.add(Conv2D(256, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.2))

    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(17, activation='sigmoid'))

    model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=[f1_score_fun,accuracy_fun])

    return model
```

```python
# saving best checkpoints 
save_best_check_point= ModelCheckpoint(filepath = 'best_model.hdf5', monitor = 'val_f1_score_fun', mode = 'max',
                                       save_best_only = True, save_weights_only = True)
```

```python
#initializing imagedatagenerator with a validation split of 0.2
image_gen_train = ImageDataGenerator(rescale = 1/255, validation_split = 0.2)

# train_set contains 80 percent of images
train_gen = image_gen_train.flow_from_dataframe(dataframe=train_csv_copy,
                                                directory ="D:/Dataset/planets_dataset/planet/planet/train-jpg",  
                                                x_col="image_name", y_col=columns, subset="training", 
                                                batch_size=200,seed=2021, shuffle=True, 
                                                class_mode="raw", target_size=(128,128))

#while val_set contains 20 percent of images 
val_gen = image_gen_train.flow_from_dataframe(dataframe=train_csv_copy,
                                                directory ="D:/Dataset/planets_dataset/planet/planet/train-jpg",  
                                                x_col="image_name", y_col=columns, subset="validation", 
                                                batch_size=200,seed=2021, shuffle=True, 
                                                class_mode="raw", target_size=(128,128))
```

```python
#setting up step size for training and validation image data
steps_for_train = int(np.ceil(train_gen.samples / train_gen.batch_size))
steps_for_val = int(np.ceil(val_gen.samples / val_gen.batch_size))
```

```python
model_testing = build_model()
#this is how our model looks 
model_testing.summary()
```

```python
%%time
#fitting the model
r = model_testing.fit(
    x=train_gen,
    steps_per_epoch=steps_for_train,
    validation_data=val_gen,
    validation_steps=steps_for_val,
    epochs=15,
    callbacks=[save_best_check_point],
    workers=-1,
    use_multiprocessing=True,
)
```

```python
#making another model with which we can make predictions 
model_submission = build_model()
```

```python
#our images don't have .jpg extensions so we need to add those extensions 
submisssion_file = pd.read_csv('D:/Dataset/planets_dataset/planet/planet/sample_submission.csv')
submisssion_file = submisssion_file.copy()
submisssion_file['image_name'] = submisssion_file['image_name'].apply(lambda x: '{}.jpg'.format(x))
submisssion_file
```

```python
#loading in the weights of the our best_model so that we can make predictions with it
model_submission.load_weights('best_model.hdf5')
```

```python
#we divide the submission file into two splits as the image names after 40669 is different 
test_df1 = submisssion_file.iloc[:40669]['image_name'].reset_index().drop('index', axis =1)
test_df1
```

```python
#initializing imagedatagenerator for the test images and also rescaling
image_gen_test = ImageDataGenerator(rescale = 1/255)


#creating a generator 
test_df_gen = image_gen_test.flow_from_dataframe(dataframe=test_df1, 
                                                directory="D:/Dataset/planets_dataset/planet/planet/test-jpg", 
                                                x_col="image_name", y_col=None, batch_size=200, 
                                                shuffle=False, class_mode=None, target_size=(128,128))
#defining step fun
step_test_size_submission = int(np.ceil(test_df_gen.samples/test_df_gen.batch_size))
```

```python
#dealing with the remaing images
test_df_gen1 = image_gen_test.flow_from_dataframe(dataframe=test_df1, 
                                                directory="D:/Dataset/planets_dataset/test-jpg-additional/test-jpg-additional", 
                                                x_col="image_name", y_col=None, batch_size=200, 
                                                shuffle=False, class_mode=None, target_size=(128,128))

step_test_size_submission2 = int(np.ceil(test_df_gen1.samples/test_df_gen1.batch_size))
```
